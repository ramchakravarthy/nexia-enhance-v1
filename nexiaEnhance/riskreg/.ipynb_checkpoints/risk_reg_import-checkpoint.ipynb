```python
def cleanse_sheet(df_temp, sheet, risk_response_source, entry_type):
    df_temp['SNO'] = np.arange(1,len(df_temp)+1)

    print("Step 1 parse done")

    df_temp = df_temp[['SNO', 'Quality objective', 'Quality Risk', 
   'Responses to address the quality risks', 'Mandatory?', 'Applicable?', 'Owner',
   'Response type', 'Response status', 'Comments', 'Review frequency']]

    print("Step 2 Filter done")

    df_1 = df_temp[['SNO', 'Quality objective', 'Quality Risk', 
   'Responses to address the quality risks']]
    df_1.fillna(method="ffill", inplace=True)

    print("Step 3 Fill na done")

    df_2 = df_temp[['SNO', 'Mandatory?', 'Applicable?', 'Owner',
   'Response type', 'Response status', 'Comments', 'Review frequency']]

    print("Step 4 second df done")

    df_temp = pd.merge(left = df_1, left_on="SNO", right = df_2, right_on="SNO", how="inner")
    del df_1, df_2

    print("Step 5 merge two dfs")

    df_temp = df_temp[df_temp['Applicable?']=="Yes"]

    print("Step 6 Filter by applicable done")
    
    df_temp.drop(columns=["SNO"], inplace=True)
    
    print("Step 7 dropped SNO done")
    
    df_temp['Quality objective catgory'] = df_temp['Quality objective'].apply(lambda x: sheet)
    
    return df_temp
```
